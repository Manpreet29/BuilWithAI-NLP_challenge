```python
# Importing the libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import datetime 
import plotly.io as pio
pio.templates.default = 'plotly_dark'
```

```python
topic = pd.read_excel('../input/hackathon/cnn_dominant_topic_date.xlsx')
topic.head()
```

```python
import datetime
def dateq(a):
    a = datetime.datetime.strptime(a, '%b %d, %Y').strftime('%d/%m/%Y')
    return a
    
dateq('May 9, 2020')
```

```python
topic.published_date = topic.published_date.apply(dateq)
```

```python
monthDict={1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
x = 'May 9, 2020'
```

```python
topic.Dominant_Topic = topic.Dominant_Topic.astype('category')
topic.published_date = pd.to_datetime(topic.published_date)
```

```python
topic.info()
```

```python
topic.sort_values(by='published_date', inplace=True)
```

```python
topic.to_excel('topic.xlsx')
```

```python
topic = topic[23:]
```

```python
dates_unique = topic.published_date.unique() 
dates_unique
```

```python
li = dates_unique
li.sort()
```

```python
li
```

```python
topic_gb = topic.groupby(['published_date'])
```

```python
dict1 = {0: 33.422, 1: 39.2308, 2: 30.132}
dict2 = {0: 42.2422, 1: 43.342, 2: 42.424}
dict3 = {0: 13.422, 1: 9.2308, 2: 20.132}
dicts = []
dicts.append(dict1)
dicts.append(dict2)
dicts.append(dict3)
dicts
```

```python
cat_unique = topic.Dominant_Topic.unique()
t_dict= []
for i in li:
    v = topic_gb.get_group(i).Dominant_Topic.value_counts().to_dict()
    #print(v)
    t_dict.append(v)
```

```python
t = pd.DataFrame(t_dict)
```

```python
t['Date'] = li
```

```python
t.replace(np.NaN, 0, inplace=True)
t.head()
```

```python
t.columns
```

```python
last_days_trend = go.Figure()
last_days_trend.add_trace(go.Scatter(x=t['Date'],y = t[0],
                                 mode='lines+markers',
                                 name='Police', 
                                 line=dict(color='darkblue', width=3)))

last_days_trend.update_traces(showlegend=False)
last_days_trend.update_layout(title={'text': "Popularity For the topic [Police]",
                                      'y':0.9,'x':0.50,
                                      'xanchor': 'center','yanchor': 'top'},
                                       xaxis_title="Date", yaxis_title="Daily Topic Count on Articles",
                                       legend_orientation="h",height=600)
last_days_trend.show()
last_days_trend.write_html("Police.html")
```
